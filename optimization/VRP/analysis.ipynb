```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('results/results.csv', sep=';')
df.head()
```

```python
# Plot
plt.figure(figsize=(8, 5))
sns.barplot(x='nr_of_clients', y='optimal_cost', hue='type', data=df, palette='viridis')

# Labels and title
plt.xlabel('Type')
plt.ylabel('Optimal Cost')
plt.title('Optimal Cost by Type and Number of Clients')
plt.legend(title="Number of Clients")

# Show the plot
plt.show()
```

```python
# Plot
plt.figure(figsize=(8, 5))
sns.barplot(x='nr_of_clients', y='duration_seconds', hue='type', data=df, palette='viridis')

# Labels and title
plt.xlabel('Type')
plt.ylabel('Duration (s)')
plt.title('Duration by Type and Number of Clients')
plt.legend(title="Number of Clients")

# Show the plot
plt.show()
```
