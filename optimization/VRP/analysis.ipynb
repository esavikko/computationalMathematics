```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('results/results.csv', sep=';')
df.head()
```

```python
# Plot
plt.figure(figsize=(8, 5))
sns.barplot(x='type', y='optimal_cost', hue='nr_of_clients', data=df, palette='viridis')

# Labels and title
plt.xlabel('Type')
plt.ylabel('Optimal Cost')
plt.title('Optimal Cost by Type and Number of Clients')
plt.legend(title="Number of Clients")

# Show the plot
plt.show()
```
